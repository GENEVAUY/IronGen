## Dividing the DataFrame into numerical and categorical

```python
categorical = df.select_dtypes(include=['object', 'category'])

numerical = df.select_dtypes(include=['int', 'float'])

print("Categorical variables:")
print(categorical.columns)

print("\nNumerical variables:")
print(numerical.columns)
```

```python
numerical
```

## Checking the correlation features inside the numerical DataFrame

```python
correlation = numerical.corr().abs()
threshold = 0.8

highly_correlated = (correlation > threshold).sum() - 1
highly_correlated = highly_correlated[highly_correlated > 0]

if not highly_correlated.empty:
    print("Highly correlated features:")
    print(highly_correlated)
else:
    print("No highly correlated features found.")
```

```python
correlation
```

## Making a heatmap to properly see the correlation

```python
plt.figure(figsize=(12, 8))
sns.heatmap (correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## checking how the graphs of each column looks like - to see the distribution and skewness

```python
from scipy import stats
import math

num_columns = numerical.shape[1]
num_rows = math.ceil(num_columns / 4)
fig, axes = plt.subplots(num_rows, 4, figsize=(16, num_rows * 3))
fig.subplots_adjust(hspace=0.5)

for i, column in enumerate(numerical.columns):
    row = i // 4
    col = i % 4
    ax = axes[row, col] if num_rows > 1 else axes[col]

    sns.histplot(numerical[column], kde=True, ax=ax)

    ax.set_title(f'Bell Curve: {column}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Density')

if num_columns % 4 != 0:
    for j in range(num_columns % 4, 4):
        fig.delaxes(axes[-1, j] if num_rows > 1 else axes[j])

plt.show()
```

## using the Standardscaler to the numerical DataFrame and graphing it

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import math

numerical_columns = numerical.columns  # Get all columns from the numerical data frame

scaler = StandardScaler()
numerical_scaled2 = pd.DataFrame(scaler.fit_transform(numerical[numerical_columns]), columns=numerical_columns)

# The 'numerical_scaled2' data frame will contain the standardized values of the numerical columns

num_plots = len(numerical_scaled2.columns)
num_rows = math.ceil(num_plots / 4)

fig, axes = plt.subplots(num_rows, 4, figsize=(16, 3*num_rows))

for i, column in enumerate(numerical_scaled2.columns):
    row = i // 4
    col = i % 4
    axes[row, col].hist(numerical_scaled2[column], bins=10)
    axes[row, col].set_title(column)

plt.tight_layout()
plt.show()
```

## using the MinMaxcaler to the numerical DataFrame and graphing it

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_scaled = pd.DataFrame(scaler.fit_transform(numerical), columns=numerical.columns)

print(numerical_scaled)
```

```python
num_plots = len(numerical_scaled.columns)
num_rows = math.ceil(num_plots / 4)

fig, axes = plt.subplots(num_rows, 4, figsize=(16, 3*num_rows))

for i, column in enumerate(numerical_scaled.columns):
    row = i // 4
    col = i % 4
    axes[row, col].hist(numerical_scaled[column], bins=10)
    axes[row, col].set_title(column)

plt.tight_layout()
plt.show()
```

## using the PowerTransformer to the numerical DataFrame to fix the skewness and graphing it 

```python
from sklearn.preprocessing import PowerTransformer

numerical_columns = numerical.columns  # Get all columns from the numerical data frame

# Apply PowerTransformer
power_transformer = PowerTransformer()
numerical_transformed = power_transformer.fit_transform(numerical[numerical_columns])
numerical_scaled3 = pd.DataFrame(numerical_transformed, columns=numerical_columns)

# The 'numerical_scaled3' data frame will contain the transformed values using PowerTransformer

# Standardize the transformed data
scaler = StandardScaler()
numerical_scaled3 = pd.DataFrame(scaler.fit_transform(numerical_scaled3[numerical_columns]), columns=numerical_columns)

# The 'numerical_scaled3' data frame will contain the standardized values of the transformed columns

num_plots = len(numerical_scaled3.columns)
num_rows = math.ceil(num_plots / 4)

fig, axes = plt.subplots(num_rows, 4, figsize=(16, 3*num_rows))

for i, column in enumerate(numerical_scaled3.columns):
    row = i // 4
    col = i % 4
    axes[row, col].hist(numerical_scaled3[column], bins=10)
    axes[row, col].set_title(column)

plt.tight_layout()
plt.show()
```
